```python
#линейная интерполяция без groupby,apply, interpolate
step = 4
for j in range(0, len(columns_to_fill)):
    N = len(df[columns_to_fill[j]])
    for i in range(0, N):
        if np.isnan(df[columns_to_fill[j]].values[i]):# and not np.isnan(v.values[i - 1]):
            start = -1
            end = -1
            for i1 in range(i-1, i-1-step,-1):
                 if not np.isnan(df[columns_to_fill[j]].values[i1]) and df['index'].values[i]==df['index'].values[i1]:
                        start = i1
                        break   
            for i1 in range(i+1, i+1+step, 1):
                 if not np.isnan(df[columns_to_fill[j]].values[i1]) and df['index'].values[i]==df['index'].values[i1]:
                        end = i1
                        break
            if (start !=-1) and (end !=-1) and (end - start <= step+1):
                y1 = df[columns_to_fill[j]].values[end]
                y0 = df[columns_to_fill[j]].values[start]
                for i1 in range(start+1, end):
                    df[columns_to_fill[j]].values[i1]=(y1-y0)*(i1 - start)/(end-start) + y0
                i = end + 1
        
```
